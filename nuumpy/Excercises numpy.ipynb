```python
# 7- what is the max and min height
print(data[:,0].min(),',',data[:,0].max())
height = data[:,0]
print('max: ', height.max())
print('min: ', height.min())
```

```python
# 8- what is the max and min weight
print(data[:,1].min(),',',data[:,1].max())
weight = data[:,1]
print('max: ', weight.max())
print('min: ', weight.min())
```

```python
print(np.min(height),np.max(height))
print(np.min(weight),np.max(weight))
```

```python
# 9- find the mean and str of height using the sample operations only
mean_h = height.sum() / height.size
print('the Mean of height:',mean_h)
print('')
str1_h = mean_h - height
squared_h = str1_h ** 2     # np.power(str1_h, 2)
sum_h = squared_h.sum()
n_h = sum_h / (height.size - 1)
std_h = np.sqrt(n_h)
print('the Standard Deviation: ',std_h)
```

```python
# 10- find the mean and str of weight using the sample operations only
mean_w = weight.sum() / weight.size
print('the Mean of weight:',mean_w)
print('')
str1_w = mean_w - weight
squared_w = str1_w ** 2  # np.power(str1_w, 2)
sum_w = squared_w.sum()
n_w = sum_w / (weight.size - 1)
std_w = np.sqrt(n_w)
print('the Standard Deviation: ',std_w)
```

```python
# 11- find the mean and str of height 
print(np.mean(height))
print()
print(np.std(height))
```

```python
# 12- find the mean and str of weight 
print(np.mean(weight))
print()
print(np.std(weight))
```

```python
# 13- find the 25th , 75th percentile and the median for height
np.percentile(height, [25,50,75]) , np.median(height)
```

```python
# 14- find the 25th , 75th percentile and the median for weight
print(np.median(weight))
print(np.percentile(weight, [25,50,75]))
```

```python
# 15- normalize height and weight
nor_h = (height - mean_h) / std_h
print(nor_h)
print()
nor_w = (weight - mean_w) / std_w
print(nor_w)
```

```python
# min max scaler
scaled_h = (height - np.min(height)) / (height.max() - height.min())
print(scaled_h)
```
