```python
fig = px.bar(data_frame=df,x='embark_town',template='plotly_dark',width=700)
fig.show()
fig = px.bar(data_frame=df,x='who',template='plotly_dark',width=700)
fig.show()
fig = px.bar(data_frame=df,x='alive',template='plotly_dark',width=700)
fig.show()
fig = px.bar(data_frame=df,x='class',template='plotly_dark',width=700)
fig.show()
fig = px.bar(data_frame=df,x='age',template='plotly_dark',width=700)
fig.show()
```

```python
app = dash.Dash(__name__)

#-----------------------------------------
app.layout = html.Div([
    html.Div([
        html.Label(['Explorty analysis for titanic Dataset']),
        dcc.Dropdown(
            id='my_dropdown',
            options=[
                      {'label': 'sex','value':'sex'},
                      {'label': 'age','value':'age'},
                      {'label': 'who','value':'who'},
                      {'label': 'class','value':'class'},
                      {'label': 'embark_town','value':'embark_town'}
                      
            ],
            value='who',
            multi=False,
            clearable=False,
            style={'width':'80%'}
        ),
    ]),
    html.Div([
        dcc.Graph(id='the_graph')
    ]),
])

#-----------------------------------------------------------------
@app.callback(
    Output(component_id='the_graph',component_property='figure'),
    [Input(component_id= 'my_dropdown', component_property='value')]
)

def update_graph(my_dropdown):
    dff = df

    piechart = px.pie(
                data_frame=dff,
                names=my_dropdown,
                hole=.3,
                template='plotly_dark'
    )
    return (piechart)

#-------------------------------------------
if __name__ == '__main__':
    app.run_server()
```
