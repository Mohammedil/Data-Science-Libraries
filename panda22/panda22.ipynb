## ***combining dataframe***
***

- pd.merge(a, b, how = 'right' , on = 'id')
- c.join(d, how= 'inner', lsuffix='_from_left', rsuffix='_from_right')
- c.append(d, ignore_index=True)
- pd.concat([c,d], axis=0) + pd.concat([c,d], axis=1)

```python

a = pd.DataFrame({'id':[1,3,5,7],
                 'grade1': [22,44,55,66],
                 'stars1':[2,4,3,1]})
a = a.set_index('id')
a
```

```python
b = pd.DataFrame({'id':[2,4,6,8],
                 'grade':[44,22,77,55],
                 'stars2':[7,5,1,2]})
b
```

```python
pd.merge(a, b, how = 'right' , on = 'id')
```

```python
pd.merge(a, b, how = 'outer' , on = 'id')
```

```python
c = pd.DataFrame({'id':[1,2,3,4],
                 'grade': [22,44,55,66],
                 'stars':[2,4,3,1]})
c
```

```python
d = pd.DataFrame({'id':[1,2,3,4],
                 'grade':[44,22,77,55],
                 'stars':[7,5,1,2]})
d
```

```python
pd.merge(right=c , left=d , how='inner', on='id')
```

```python
c.join(d, how= 'inner', lsuffix='_from_left', rsuffix='_from_right')
```

```python
c = c.set_index('id')
d = d.set_index('id')
```

```python
pd.concat([c,d])
```

```python
pd.concat([c,d], keys = ['x','y'])
```

```python
pd.concat([c,d], keys = ['x','y']).reset_index()
```

```python
pd.concat([c,d], axis=0)
```

```python
pd.concat([c,d], axis=1)
```

```python
c.append(d, ignore_index=True)
```

##### ***filter***

```python
def filter_func(dataframe):
    return dataframe['Data2'].mean() < 5
   
```

```python
dataa.groupby('key').filter(filter_func)
```

```python
def ratio(dataframe):
    dataframe['ratio'] = dataframe['Data1'] / dataframe['Data2']
    return dataframe
```

```python
dataa.groupby('key').apply(ratio)
```

```python
def ratio(dataframe):
    dataframe['ratio'] = dataframe['Data1'] + dataframe['Data2']
    return dataframe
```

```python
dataa.groupby('key').apply(ratio)
```

```python
def trans(column):
   
    return  (column - column.mean()) / column.std()
```

```python
dataa.groupby('key').transform(trans)
```

```python
def add_5(column):
   
    return  column + 5
```

```python
dataa.groupby('key').transform(add_5)
```

## ***pivot table***
***

```python
tit = sns.load_dataset('titanic')
tit
```

```python
tit['sex'].value_counts()
```

```python
tit['who'].value_counts()
```

```python
tit.pivot_table( index= 'who', columns= 'class' ,values= 'age' ,aggfunc= 'mean' )
```

```python
tit.pivot_table( index= 'sex',values= 'survived' ,aggfunc= 'mean', columns= 'class' )
```

```python
tit.pivot_table( index= 'sex',values= ['survived', 'age' ] ,aggfunc= 'count', columns= 'class' )
```

```python
tit.pivot_table( index= ['sex', 'who'],values= ['survived', 'age' ] ,aggfunc= { 'survived':'median','age': 'mean'}, columns= 'class' )
```
