```python
# Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***
## ***line plot***
***

```python
fl = sns.load_dataset('flights')
fl
```

```python
flf = fl[fl['month'] == 'Feb']
flf
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,8))
sns.lineplot(x= 'year', y= 'passengers', data=fl, hue='month', size= 'month')
plt.title('years vs  passengers')
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
plt.figure(figsize= (8,6))
sns.lineplot(data= fmri, x= 'timepoint', y= 'signal', hue='event',style='region')
```

```python
# relplot
sns.relplot(data= fmri, x= 'timepoint', y= 'signal', hue='event',style='region', kind='line', col= 'region', row= 'event')
```

___
## ***scatter plots***
___

```python
sns.set_style('dark')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data= tips, x= 'total_bill', y='tip')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data= tips, x= 'total_bill', y='tip', hue='sex', style='smoker',size='size', sizes=(30,200))
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data= tips, x= 'total_bill', y='tip', hue='sex' ,size='size', sizes=(30,200), palette='plasma') # palette matplotlib
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
sns.scatterplot(data= tips, x= 'total_bill', y='tip', hue='size' ,size='size', sizes=(30,200), palette='Purples')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data= tips, x= 'total_bill', y='tip', hue='time', markers={'Lunch':"s", 'Dinner':"D"}) # markers matplotlib
```

```python

sns.relplot(data= tips, x= 'total_bill', y='tip', hue='time', style='sex' , kind='scatter' ,col='sex', row= 'day')
```

## ***Regression plot***
***

```python
tips
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x='size', y='tip', data=tips)
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x='size', y='tip', data=tips, x_jitter=0.1)
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x='size', y='tip', data=tips, x_estimator=np.mean)
```

```python
plt.figure(figsize=(8,5))
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', col='smoker', row= 'sex',truncate=False , scatter_kws={'s':100,'edgecolor':'black'}, line_kws={'lw':6,"ls":'--'})
```

```python
tips['smoker'] = tips['smoker'] == 'Yes'
```

```python
plt.figure(figsize=(8,5))
sns.lmplot(x='total_bill', y='smoker', data=tips)
```

```python
pen = sns.load_dataset('penguins')
pen.head()
```

##### sns.histplot

```python
sns.histplot(pen, x='flipper_length_mm', bins=20)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(pen, x='flipper_length_mm', bins=20, stat= 'density')
```

```python
plt.figure(figsize=(8,6))
sns.histplot(pen, x='species', stat= 'density', shrink=0.8)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(pen, x='flipper_length_mm', stat= 'density', shrink=0.9, hue='species')
```

```python
plt.figure(figsize=(8,6))
sns.histplot(pen, x='flipper_length_mm', stat= 'density', hue='species', element='step')
```

```python
sns.histplot(pen, x='flipper_length_mm', bins=20, kde= True)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(pen, x='flipper_length_mm', stat= 'density', shrink=0.9, hue='species', kde=True)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(pen, y='flipper_length_mm', stat= 'density', shrink=0.9, hue='species', kde=True)
```
